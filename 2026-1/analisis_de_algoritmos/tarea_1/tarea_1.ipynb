5. Determinar la función $T(n)$ en el mejor y el peor de los casos del algoritmo de ordenamiento por selección.
<br>Ya que  el algoritmo de ordenamiento por selección es lo que llamamos un "algoritmo ciego", el T(n) es el mismo para el mejor caso y el peor caso (además del promedio). Por lo tanto solo calculamos $T(n)$ una vez.

    def selection_sort_muestra(arr):
        
        n = len(arr) # una vez, c_1
        
        for i in range(n - 1): # c2

            min_idx = i # c2
            
            for j in range(i + 1, n): # c3
                if arr[j] < arr[min_idx]: # c3
                
                    min_idx = j # c3
            
            arr[i], arr[min_idx] = arr[min_idx], arr[i] # c_2

Para $c_{1}$, una asignación simple que se ejecuta una vez, su coeficiente es 1.
Para $c_{2}$, en el mismo código podemos observar que se ejecuta $n-1$ veces.
$c_{3}$ es donde se vuelve más complicado.

    for j in range(i + 1, n):
        if arr[j] < arr[min_idx]:

            min_idx = j

El número de veces que el loop se ejecuta depende de $i$.<br>
Si $i = 0$ el bucle va desde 1 a $n - 1$.<br>
Si $i = 1$ el bucle va desde 2 a $n - 1$.<br>
<br>
Para llegar al final del bucle entonces $i + 1 = n-1 \rightarrow i = n - 2$<br>
Entonces se forma una secuencia: $n - 1, n - 2, (...)$ que no es más que un número triangular en orden inverso.<br>
Para calcular el número de veces que se ejecute el bucle sustituimos $n-1$ en la fórmula del número triangular:<br>
$$T_k = \frac{k(k+1)}{2} \, \rightarrow \, T_{n-1} = \frac{(n-1)(n)}{2}$$
Entonces, el coeficiente del bloque $c_{3}$ es igual a $$ T_{n-1} = \frac{n^{2}-n}{2} $$
Por lo tanto $$ T(n) = c_{1} + c_{2}(n-1) + c_{3}\left(\frac{n^{2}-n}{2}\right)$$
